```python
with open("i13") as f:
    lines = [line.strip() for line in f.readlines()]
    

import re
As = [tuple(map(int,re.findall("\d+", line))) for line in lines[::4]]
Bs = [tuple(map(int,re.findall("\d+", line))) for line in lines[1::4]]
Ps = [tuple(map(int,re.findall("\d+", line))) for line in lines[2::4]]
A,B,P = As[0],Bs[0],Ps[0]
print(A,B,P)
import numpy as np
a = np.column_stack([A,B])
a
```

```python
def get_ab(A,B,P):
    a = np.column_stack((A,B))
    P = np.array(P)
    ia = np.linalg.inv(a)
    uv = np.dot(ia, P)
    # print(np.linalg.det(a))
    return uv

def get_pos(A,B, presses):
    return np.dot(np.column_stack([A,B]), presses)

def winnable(uv):
    
    return abs(sum(uv) - round(sum(uv))) < 0.0000000001
uv = winnable(get_ab(A,B,P))
uv, sum(get_ab(A,B,P))

uv
```

```python
tokens = 0
wins = 0
winners = []
for A,B,P in zip(As,Bs,Ps):
    # Px,Py = P
    # P = [Px+10000000000000,Py+10000000000000]
    px,py = get_ab(A,B,P)
    px,py = int(px), int(py)
    close = get_pos(A,B,[px,py])
    ab = get_ab(A,B,close)
    if winnable(ab):
        a,b=ab 
        tokens+=round(3*(a+px)+ py+b)
        wins += 1
        
    winners.append((A,B,P))
    print(A,B, winnable(ab), wins, tokens,  ab)
```
