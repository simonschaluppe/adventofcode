```python
# get the path of the guard

with open("i6.txt") as f:
    area = [[x for x in line.strip()] for line in f.readlines()]
    
def show(area=area, empty=".", crop=None, path=None):
    if crop is None: crop = len(area)-1 
    for row in area[:crop]:
        print(''.join(row[:crop]).replace(".", empty))

size = len(area)  
ix, iy = 45,42
direction = 0
dirs = [(-1,0),(0,1),(1,0),(0,-1)]
arrows = [(-1,0),(0,1),(1,0),(0,-1)]
def get_guard_path(ix=45,iy=42,d=0,area=area,mark=">"):
    """returns the path a guard takes for a given area 
    as well as True/False if the path loops, meaning a tile is exited twice in the same direction 
    meaning (x,y,direction) already in path 
    """
    gx, gy, direction = ix,iy,d
    guard_path = set()
    guard_path.add((ix,iy,direction))
    while not any([gx==0, gy==0, gx==size-1, gy==size-1]):
        dx, dy = dirs[direction]
        nx, ny = gx+dx, gy+dy
        if area[nx][ny] == "#":
            direction = (direction+1)%4
        elif (nx,ny,direction) in guard_path: # its already closed
            return guard_path, True
        else:
            guard_path.add((nx,ny,direction))
            gx, gy = nx, ny
    return guard_path, False

guard_path, looping = get_guard_path()
visited = set([(nx,ny) for nx,ny,_ in guard_path])
len(visited), len(guard_path), looping
```

```python
obstacles = set()
loops = []
escapes = []

guard_path.remove((45,42,0))
for path in guard_path:
    x,y,_= path
    obstacled_area = [line.copy() for line in area]
    obstacled_area[x][y] = "#"
    alternative_path, is_looping = get_guard_path(area=obstacled_area) # the initial mistake was to check the guard path from the obstacle rather than the beginning
    # but an obstacle can prevent the guard from reaching its later obstacled path from a different direction
    #alternative_path, is_looping = get_guard_path(*path, obstacled_area) # the initial mistake was
    if is_looping:
        loops.append(alternative_path)
        obstacles.add((x,y))
    else:
        escapes.append(alternative_path)
```
