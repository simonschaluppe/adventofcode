```python
from aocutils.utils import show
city = [
"............",
"........0...",
".....0......",
".......0....",
"....0.......",
"......A.....",
"............",
"............",
"........A...",
".........A..",
"............",
"............",]

city = [list(l) for l in city]
show(city, use_emojis=True)
```

```python

def get_towers(city):
    towers = dict()
    for i in range(len(city)):
        for j in range(len(city[0])):
            location = city[i][j]
            if location =='.':continue
            if location in towers:
                towers[location].append((i,j))
            else:
                towers[location] = [(i,j)]
    return towers

towers = get_towers(city)
towers
```

```python
from itertools import combinations

def next_antinode(p,d,bounds):
    (px, py), (dx, dy) = p, d
    nx, ny = px+dx, py+dy
    if 0<=nx<bounds and 0<=ny<bounds:
        return (nx,ny)
    else: return None

def find_antinodes(pairs, bounds, opposite_only=True):
    antinodes = set()
    for pair in pairs:
        (x1,y1),(x2,y2) = pair
        dx, dy =  x2-x1,y2-y1
        if opposite_only:
            if n:=next_antinode((x1,y1),(2*dx,2*dy), bounds): antinodes.add(n)
            if n:=next_antinode((x2,y2),(-2*dx,-2*dy), bounds): antinodes.add(n)
        else:
            n = next_antinode((x1,y1),(dx,dy), bounds)
            while n: 
                antinodes.add(n)
                n = next_antinode(n,(dx,dy), bounds)
            n = next_antinode((x2,y2),(-dx,-dy), bounds)
            while n: 
                antinodes.add(n)
                n = next_antinode(n,(-dx,-dy), bounds)
    return antinodes

def get_tower_pairs(list_of_towers):
    return combinations(list_of_towers,2)

pairs = set(get_tower_pairs(get_towers(city)["A"]))
antinodes = set(find_antinodes(pairs, len(city)))
antinodes
    
```

```python
from aocutils.utils import mark_on_map, replace_on_map

def analyze_city(city, opposites_only=True, emojis=True):
    towers = get_towers(city)
    antinodes = set()
    for frequency in towers.values():
        pairs = get_tower_pairs(frequency)
        antinodes.update(find_antinodes(pairs, len(city), opposite_only=opposites_only))

    node_map = mark_on_map(antinodes, city, "#")
    print("Antinodes: ", len(antinodes))
    show(node_map, use_emojis=emojis)
    return len(antinodes)

analyze_city(city, emojis=False)
```
