```python
with open("i7.txt") as f:
    lines = [line.strip() for line in f.readlines()]
    
results = [int(x.split(":")[0]) for x in lines]
operands =[[int(o.strip()) for o in ops.split(":")[1].split(" ")[1:]] for ops in lines] 
size = len(operands)
size, lines[:5], results[:5], operands[:5]
```

how  to create all possible results?
- loop through the operands
- branch * and + each time
- if > result at any time -> return false


```python
test_result, test_ops = results[1], operands[1]

possible_results = [test_ops[0]]
    
def possibilites(interim_results, ops, expected_result):
    new_results = []
    for possible_result in interim_results:
        mul_result = possible_result * ops[0]
        add_result = possible_result + ops[0]
        if mul_result <= expected_result:
            new_results.append(mul_result)
        if add_result <= expected_result:
            new_results.append(add_result)
    
    if not new_results: # overshoot #should be possible to do
        return 0
    if len(ops) == 1: # last operator
        print(expected_result,new_results)
        return len([r for r in new_results if r == expected_result])
    else:
        return possibilites(new_results, ops[1:], expected_result)
    
possibilites(possible_results, test_ops[1:], test_result)

```

```python
possibly_true = []
# results = [190,3267,292]
# operands = [[10,19],[81,40,27],[11,6,16,20]]

for result, ops in zip(results, operands):
    possibilities_result = possibilites([ops[0]], ops[1:], result)
    possibly_true.append(possibilities_result)
max(possibly_true)

# print(possibly_true)
# sum(1 for p in possibly_true if p)
```

```python
test_result, test_ops = results[1], operands[1]

possible_results = [test_ops[0]]
    
def possibilites(interim_results, ops, expected_result, verbose=False):
    new_results = []
    if verbose: print(interim_results, ops[0])
    for possible_result in interim_results:
        mul_result = possible_result * ops[0]
        add_result = possible_result + ops[0]
        concat_result = possible_result * 10**(len(str(ops[0]))) + ops[0]
        if mul_result <= expected_result:
            new_results.append(mul_result)
        if add_result <= expected_result:
            new_results.append(add_result)
        if concat_result <= expected_result:
            new_results.append(concat_result)
    
    if verbose: print(expected_result,new_results)
    if not new_results: # overshoot #should be possible to do
        return 0
    if len(ops) == 1: # last operator
        return len([r for r in new_results if r == expected_result])
    else:
        return possibilites(new_results, ops[1:], expected_result, verbose=verbose)
    
possibilites(possible_results, test_ops[1:], test_result, verbose=True)

```
