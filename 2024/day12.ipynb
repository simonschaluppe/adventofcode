```python
from aocutils.utils import *

field_map = input_as_nested_list("i12")
# field_map = [list(l) for l in """RRRRIICCFF
# RRRRIICCCF
# VVRRRCCFFF
# VVRCCCJFFF
# VVVVCJJCFE
# VVIVCCJJEE
# VVIIICJJEE
# MIIIIIJJEE
# MIIISIJEEE
# MMMISSJEEE""".split("\n")]
print(len(field_map))
show(field_map, use_emojis=True)
```

```python

field_catalog = {}

running_id = 0
    
def catalogue(position):
    global running_id
    x,y = position
    if (x,y) in field_catalog: return
    croptype = field_map[x][y]
    neighbors = get_neighbors((x,y), field_map)
    already_in_catalog = [n in field_catalog and field_catalog[n][0]==croptype for n in neighbors]
    if not any(already_in_catalog):
        field_id = running_id
        running_id += 1
    else:
        for n, is_there in zip(neighbors, already_in_catalog):
            if is_there:
                field_id = field_catalog[n][1]
    field_catalog[(x,y)] = (croptype, field_id)
    for nx,ny in neighbors:
        if field_map[nx][ny]==croptype: catalogue((nx,ny))
            

for x in range(len(field_map)):
    for y in range(len(field_map)):
        catalogue((x,y))
```

```python
fields = {}

iis = ["^",">", "v", "<"]

class Side:
    normal:tuple[int,int]
    members:list
    
class Field:
    def __init__(self,ID, crop):
        self.ID = ID
        self.crop = crop
        self.plots = set()
        
    def get_sides(self):
        self.sidepart_catalog = {}
        self.side_nr = 1
        self.borders = self.border_plots()
        for position in self.borders:
            for normal in self.border_normals(position):    
                self.catalogue_side(position, normal)
        return len(set(ID for ID in self.sidepart_catalog.values()))
    
    def catalogue_side(self, position, i):
        x,y = position
        if (x,y,iis[i]) in self.sidepart_catalog: return
        for nex,ney in [DIRS_HORIVERT[(i-1)%4], DIRS_HORIVERT[(i+1)%4]]:
            if (x+nex,y+ney, iis[i]) in self.sidepart_catalog:
                snr = self.sidepart_catalog[(x+nex,y+ney, iis[i])]
                break
        else: 
            snr = self.side_nr
            self.side_nr += 1
        self.sidepart_catalog[(x,y, iis[i])] = snr
        for nex,ney in [DIRS_HORIVERT[(i-1)%4], DIRS_HORIVERT[(i+1)%4]]:
            if all([i in self.border_normals((x+nex,y+ney)),
                    (x+nex, y+ney) in self.borders,
                    not (x+nex,y+ney,iis[i]) in self.sidepart_catalog]):  
                self.catalogue_side((x+nex,y+ney), i)
        
                    
    def border_normals(self, plot):
        d =[]
        px,py = plot
        for i,(dx,dy) in enumerate(DIRS_HORIVERT):
            if not (px+dx, py+dy) in self.plots: d.append(i) 
        return d
            
    def border_plots(self):
        return [p for p in self.plots if 4> len([n for n in get_neighbors(p, field_map) if  n in self.plots])]
        
    def border_length(self):
        l = 0
        for plot in self.border_plots():
            l+= 4 - sum(1 for n in get_neighbors(plot, field_map) if n in self.plots)
        return l
    
    def area(self):
        return len(self.plots)
    
    def fence_cost(self):
        return self.border_length()*self.area()
    
    def discounted(self):
        return (self.get_sides())*self.area()
        
    def __repr__(self):
        return f"Field({self.ID:3}, {EMOJI_MAPPING[self.crop]}, area: {self.area():3}, \
border: {self.border_length():4}, sides={self.get_sides():4},{self.border_length()>=self.get_sides()>=4} \
    cost:{self.fence_cost():6}, discounted: {self.discounted():6} {self.fence_cost()>=self.discounted()}"
        
for plot, (crop, ID) in field_catalog.items():
    if ID not in fields:
        field = Field(ID, crop)
        field.plots.add(plot)
        fields[ID] = field
    else: fields[ID].plots.add(plot)
    
```
