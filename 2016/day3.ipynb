```python
import re


with open("i3") as f:
    triangles = [tuple(map(int,re.findall("\d+", t.strip()))) for t in f.readlines()]
    
    
max(triangles[0]), triangles[0]
```

```python
def check(triangles):
    possible = 0
    for sides in triangles:
        m = max(sides)
        if m < sum(sides)-m:
            possible +=1
    return possible
        
check(triangles)
```

```python
transposed = []
for t1,t2,t3 in zip(triangles[::3], triangles[1::3], triangles[2::3]):
    v11,v21,v31 = t1
    v12,v22,v32 = t2
    v13,v23,v33 = t3
    transposed.append((v11,v12,v13))
    transposed.append((v21,v22,v23))
    transposed.append((v31,v32,v33))
  
check(transposed)  
```
