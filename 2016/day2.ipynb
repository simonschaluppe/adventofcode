```python
keypad = ["123","456","789"]


pos = (1,1)
dirs = {"U":(-1,0),"L":(0,-1),"D":(1,0),"R":(0,1)}

def move(pos, d):
    (px,py), (dx, dy) = pos, dirs[d]
    # print(keypad[px][py], d, end=" ")
    return (max(0,min(2,px+dx)),max(0,min(2,py+dy)))

with open("i2") as f:
    numbers = [line.strip() for line in f.readlines()]


code = []
for line in numbers:
    for d in line:
        pos = move(pos,d)
    code.append(pos)
    
print(*[keypad[r][c] for r,c in code])
```

```python
keypad2 = """.......
...1...
..234..
.56789.
..ABC..
...D...
.......""".split("\n")

pos=(3,1)

def move(pos, d):
    (px,py), (dx, dy) = pos, dirs[d]
    if keypad2[px+dx][py+dy] == ".":
        return (px,py)
    return (px+dx, py+dy)

with open("i2") as f:
    numbers = [line.strip() for line in f.readlines()]


code = []
for line in numbers:
    for d in line:
        pos = move(pos,d)
    code.append(pos)
    
print(*[keypad2[r][c] for r,c in code])
```
