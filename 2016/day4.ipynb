```python
with open("i4") as f:
    lines = [line.strip() for line in f.readlines()]
    
import re
from collections import Counter
from operator import itemgetter

alphabet = list("abcdefghijklmnopqrstuvwxyz")
print(len(alphabet))
def decrypt(string, shifts):
    n = []
    for c in string:
        if c==" ":
            n.append(c)
            continue
        c = n.append(alphabet[(ord(c)-97+shifts)%26])
    return "".join(n)

decrypt("aczupnetwp dnlgpyrpc sfye dstaatyr",27)
```

```python

s = 0
for line in lines:
    checksum = re.findall("\[(\w+)\]", line)[0]
    ID = int(re.findall("\d+", line)[0])
    l = len(checksum)
    encrypted = line[:-10].replace("-","")
    decryptable = line[:-10].replace("-"," ").strip()

    counter = sorted(Counter(encrypted).most_common(), key=itemgetter(0))
    counter = sorted(counter, key=itemgetter(1), reverse=True)
    
    occurences = "".join([c[0] for c in counter[:l]])
    match = occurences == checksum
    if match:
        s += int(ID)
    print(ID, decrypt(decryptable, ID%26), checksum, occurences, s, match, encrypted)
    
```
