```python
with open("i12") as f:
    instructions = [l.strip().split(" ") for l in f.readlines()]

register = {"a":0,"b":0,"c":1,"d":0}
line_nr =0
while line_nr < len(instructions):

    nextline_increment = 1
    command, *vars = instructions[line_nr]
    # print(line_nr, command, *vars)
    match command:
        case "cpy":
            if vars[0] in register:
                register[vars[1]] = register[vars[0]]
            else: register[vars[1]] = int(vars[0])
        case "inc":
            register[vars[0]] += 1
        case "dec":
            register[vars[0]] -= 1
        case "jnz":
            if not any([vars[0] == "0", vars[0] in register and register[vars[0]] == 0]): 
                nextline_increment = int(vars[1])
    # print(line_nr,nextline_increment, register)
    line_nr += nextline_increment
    
print(register)
    
                
```
