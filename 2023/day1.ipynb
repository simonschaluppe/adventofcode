```python
import re

def solution1():
    with open("i1.txt") as f:
        lines = [line.strip() for line in f.readlines()]
    s = 0
    for line in lines:
        digits = re.findall("\d", line)
        s += int(digits[0] + digits[-1])
        
    return s

solution1()
```
