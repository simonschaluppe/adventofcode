```python
import aocutils.utils as aoc

pipes = aoc.input_as_nested_list("i10")
start = None
for x, pipe in enumerate(pipes):
    for y, c in enumerate(pipe):
        if c=="S": start = (x,y)

print(start)
neighbors = {"F":((1,0),(0,1)),
          "7":((0,-1),(1,0)),
          "L":((0,1),(-1,0)),
          "J":((-1,0),(0,-1)),
          "S":((-1,0),(0,-1)),
          "|":((1,0),(-1,0)),
          "-":((0,-1),(0,1))}
loop = {}
visited = []
from collections import deque
unexplored = deque([start])
while unexplored:
    x,y = unexplored.pop()
    # print(x,y)
    pipetype = pipes[x][y]
    dirs = neighbors[pipetype]
    (dx1, dy1), (dx2,dy2) = dirs
    
    visited.append((x,y))
    link1 = (x+dx1,y+dy1)
    link2 = (x+dx2,y+dy2)
    if not link1 in visited and not link1 in loop:
        loop[(x,y)] = link1
        unexplored.appendleft(link1)
        continue
    if not link2 in visited and not link2 in loop:
        loop[(x,y)] = link2
        unexplored.append(link2)
        
        
```
