```python
def possible_arrangements(records:list[str], pattern):
    if "?" not in records[0]: return records
    newrecords = []
    hash_occurrences = pattern.split(",")
    total_hash = sum(int(h) for h in hash_occurrences)
    total_dots = len(records[0]) - total_hash
    for record in records:
        if record.count("#") < total_hash:
            newrecords.append(record.replace("?", "#", 1))
        if record.count(".") < total_dots:
            newrecords.append(record.replace("?", ".", 1))
    if newrecords:
        return possible_arrangements(newrecords, pattern)
    else: return records
```

```python
possible_arrangements(["?#.#.#..???"], "3,2,1")
```

```python
import re
def list_matches(all_possible_arrangements, pattern="3,2,1"):
    hash_occurrences = pattern.split(",")
    re_pattern = "^\.*#{"+"}\.+#{".join(hash_occurrences)+"}\.*$"
    matches = [re.findall(re_pattern, arrangement) for arrangement in all_possible_arrangements]
    return [match[0] for match in matches if match]
    
list_matches(possible_arrangements(["?###???..?????"], "3,2,1"), pattern="3,2,1")
```

```python
total_valid_arrangements = 0
with open("i12.txt") as f:
    for i, line in enumerate(f.readlines()):
        record, pattern = line.strip().split(" ")
        matches = list_matches(possible_arrangements([record], pattern), pattern)
        print(f"Row {i:4}: Found {len(matches):3} | {record:>20} | {pattern:12} {matches}")
        total_valid_arrangements += len(matches)
        
total_valid_arrangements
```
