```python
with open("i9") as f:
    sequences = [list(map(int, line.strip().split(" "))) for line in f.readlines()]
    
def extrapolate(sequence, trail = None, i=-1):
    if trail is None: trail = [sequence[i]]
    else: trail.append(sequence[i])
    # print(sequence, trail)
    sequence_below = [b-a for a,b in zip(sequence[:-1], sequence[1:])]
    
    if sequence_below[-1] == 0 and sequence_below[0] == 0:
        if sum(sequence_below) == 0:
            if i==-1:
                return sum(trail)
            if i==0:
                r = 0
                for x in reversed(trail):
                    # print(r,x)
                    r = x - r
                return r
    else: 
        # print(*sequence_below)
        return extrapolate(sequence_below, trail, i=i)
    
```

```python
s = 0
s2 = 0
for sequence in sequences:
    s+= extrapolate(sequence)
    s2 += extrapolate(sequence, i=0)

print(s, s2)        
```
