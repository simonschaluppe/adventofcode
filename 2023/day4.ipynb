```python
import re

winning_numbers = []
numbers_you_have = []
with open("i4.txt") as f:
    for line in f.readlines():
        matched = re.search(r"Card +\d+: ([\d\s]+)\|([\d\s]+)", line)
        if matched:
            winning_numbers.append(set(int(num) for num in matched.group(1).strip().split()))
            numbers_you_have.append([int(num) for num in matched.group(2).strip().split()])
        else: print(line)

wins, yours = winning_numbers[6], numbers_you_have[6]
print(wins, yours)
[win for win in yours if win in wins]
```

```python

points = 0
for wins, yours in zip(winning_numbers,numbers_you_have):
    winners = [num for num in yours if num in wins]
    #winners = wins & set(yours)
    if winners:
        p = 2**(len(winners)-1)
    else: p = 0
    print(f"{p:4} {wins} {yours}")
    points += p
    
points
```

```python
# too low: 10414




def total_copies(i, cards):
    """how many total copies a card will produce by adding however many its copies will produce"""
    initial = cards[i]
    if initial == 0: return 0 # just the scratch card
    else: return initial + sum(cards[i+1:i+initial+1])
        
    
def solve2():
    """no recursion needed if we start from the last cards, which dont explode
    with each earlier card, we simply need to sum up the final copies of their copies
    which are final because they were calced before and already terminated
    """
    copies = []
    for i, (wins, yours) in enumerate(zip(winning_numbers,numbers_you_have)):
        winners = [num for num in yours if num in wins]
        #winners = wins & set(yours)
        copies.append(len(winners))
        
    for i in range(200,-1,-1):
        # print(i, copies[i])
        copies[i] = total_copies(i, copies)
            
    return sum(copies)+201

solve2()
```
